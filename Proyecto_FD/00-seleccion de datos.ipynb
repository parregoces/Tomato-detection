# Selección de datos
Este script se utiliza para seleccionar de manera aleatoria una cantidad de datos de imagenes y etiquetas de la carpeta principal Tomato_300x300 que se descarga de la siguiente ruta  https://rdm.inesctec.pt/dataset/ii-2021-001. Esto para facilitar las tareas de entrenamiento dependiendo de su capacidad de computo puede seleccionar la cantidad deseada para entrenamiento y validación:
explo_img y explo_xml y para pruebas test_img, test_xml.

Para este ejercicio se seleccionaron 300 datos para entrenar y v


```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import os
import glob
import xml.etree.ElementTree as ET
import pandas as pd
import cv2
import random
import shutil
%matplotlib inline
%load_ext tensorboard
tf.__version__
```

```python
# Seleccionar imágenes de manera aleatoria para explorar.

def seleccionar_aleatorios(carpetas_imagenes, carpetas_xml, cantidad_seleccion):
    imagenes = os.listdir(carpetas_imagenes)
    random.shuffle(imagenes)
    seleccion = imagenes[:cantidad_seleccion]

    for imagen in seleccion:
        # Copiar la imagen
        shutil.copy(os.path.join(carpetas_imagenes, imagen), 'explo_img')

        # Obtener todos los archivos XML correspondientes a la imagen
        nombre_base = imagen.split('.')[0]
        archivos_xml = [archivo for archivo in os.listdir(carpetas_xml) if archivo.startswith(nombre_base)]

        # Seleccionar aleatoriamente un archivo XML de la lista
        archivo_xml_seleccionado = random.choice(archivos_xml)

        # Copiar el archivo XML seleccionado
        shutil.copy(os.path.join(carpetas_xml, archivo_xml_seleccionado), 'explo_xml')
```

```python
# Directorios de las carpetas de imágenes y XML
carpetas_imagenes = 'Tomato_300x300/images/'
carpetas_xml = 'Tomato_300x300/Annotations/'

# Crear directorios para guardar la selección
if not os.path.exists('explo_img'):
    os.makedirs('explo_img')
if not os.path.exists('explo_xml'):
    os.makedirs('explo_xml')

# Seleccionar aleatoriamente las imágenes y sus XML correspondientes
seleccionar_aleatorios(carpetas_imagenes, carpetas_xml, 300)
```

```python
def seleccionar_aleatorios(carpetas_imagenes, carpetas_xml, cantidad):
    # Obtener una lista de todos los archivos en las carpetas de imágenes
    archivos_imagenes = os.listdir(carpetas_imagenes)

    # Obtener una lista de todos los archivos en las carpetas de XML
    archivos_xml = os.listdir(carpetas_xml)

    # Verificar si la carpeta de selección aleatoria ya existe
    if not os.path.exists('test_img'):
        os.makedirs('test_img')
    if not os.path.exists('test_xml'):
        os.makedirs('test_xml')

    # Seleccionar aleatoriamente las imágenes y sus XML correspondientes
    seleccionados = random.sample(range(len(archivos_imagenes)), cantidad)
    for indice in seleccionados:
        nombre_imagen = archivos_imagenes[indice]
        nombre_xml = archivos_xml[indice]

        # Verificar si la imagen ya existe en la carpeta 'explo_img'
        if nombre_imagen not in os.listdir('explo_img'):
            # Copiar la imagen a la carpeta de selección aleatoria
            shutil.copy(os.path.join(carpetas_imagenes, nombre_imagen), 'test_img')

            # Copiar el XML correspondiente a la carpeta de selección aleatoria
            shutil.copy(os.path.join(carpetas_xml, nombre_xml), 'test_xml')

# Directorios de las carpetas de imágenes y XML
carpetas_imagenes = 'Tomato_300x300/images/'
carpetas_xml = 'Tomato_300x300/Annotations/'

# Seleccionar aleatoriamente las imágenes y sus XML correspondientes
seleccionar_aleatorios(carpetas_imagenes, carpetas_xml, 20)
```
